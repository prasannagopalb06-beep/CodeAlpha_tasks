```python
!pip install gradio scikit-learn
import gradio as gr
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity



questions = [
    "What courses are offered in the college?",
    "What is the admission process?",
    "Is hostel facility available?",
    "Are there separate hostels for boys and girls?",
    "Is food provided in the hostel?",
    "How is the hostel food quality?",
    "Is there a college canteen?",
    "What type of food is available in the canteen?",
    "Is vegetarian food available in the canteen?",
    "What are the canteen working hours?",
    "What is the fee structure?",
    "Does the college provide placement support?",
    "Is the college affiliated to any university?",
    "What are the college working hours?",
    "Is there a library facility?",
    "Does the college provide scholarships?",
    "Is transportation facility available?",
    "Is there a medical facility in the college?",
    "Are sports facilities available?",
    "How can I contact the college office?"

]

answers = [
    "The college offers UG and PG courses in BE AND BTECH Courses",
    "Admissions are based on councelling and Management as per university guidelines.",
    "Yes, hostel facilities are available for students.",
    "Yes, there are separate hostel facilities for boys and girls.",
    "Yes, food is provided to hostel students on a daily basis.",
    "The hostel provides hygienic and nutritious food prepared under supervision.",
    "Yes, the college has a well-maintained canteen inside the campus.",
    "The canteen provides South Indian and North Indian food items along with snacks.",
    "Yes, vegetarian food is available in the college canteen.",
    "The canteen is open from 8:30 AM to 5:00 PM on working days.",
    "The fee structure depends on the course and category of the student.",
    "Yes, the college has a dedicated placement cell for training and placements.",
    "Yes, the college is affiliated with the Anna university.",
    "The college working hours are from 9:00 AM to 4:30 PM.",
    "Yes, the college has a well-stocked library with digital and printed resources.",
    "Yes, government and private scholarships are available for eligible students.",
    "Yes, transportation facilities like college buses are available.",
    "Yes, basic medical facilities and first aid are available in the campus.",
    "Yes, the college provides sports facilities for indoor and outdoor games.",
    "You can contact the college office through the official phone number or email."
]



vectorizer=TfidfVectorizer()
faq_vectors=vectorizer.fit_transform(questions)

def college_chatbot(user_question):
  user_vec=vectorizer.transform([user_question])
  similarity=cosine_similarity(user_vec,faq_vectors)
  best_match=similarity.argmax()
  score=similarity[0][best_match]

  if score<0.2:
    return "sorry,I couldn't understand your question.please try again"
  else:
    return answers[best_match]

demo=gr.Interface(
    fn=college_chatbot,
    inputs=gr.Textbox(label="Ask your college related question"),
    outputs=gr.Textbox(label="Chatbot Answer"),
    title="College FAQ Chatbot (AI Based)",
    description="This chatbot answer common college related questions using NLP and AI."

)

demo.launch(share=True)
```
